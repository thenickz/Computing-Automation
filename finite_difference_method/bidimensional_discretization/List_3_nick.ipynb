```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def analytic_function():
    x = np.linspace(0, 2, 10000) # creating the x axis
    y = np.exp(x**4/4 - 1.5*x) # calculating y(x)
    plt.plot(x, y, 'r--', label = 'Analytic') # plotting
```

```python
def numeric_function(n):
    # defining range
    start = 0
    final = 2
    '''if n%2 == float:
        n += 1'''
    # defining scope
    h = (final-start)/n
    
    # creating the x axis
    x = np.linspace(start , final, n)
    
    # creating matrix & vector
    A = np.zeros([n, n])
    b = np.zeros(n)
    y = np.zeros(n)
    for i in range(1, n-1):
        A[i, i-1] = -1
        A[i, i+1] = 1
    # defining knowns points
    y[:] = 1
    A[0, 0] = 1
    A[-1, -1] = 1
    b[-1] = 2.72
    b[0] = 1

    # setting the A matrix
    for j in range(1000):
        for i in range(1, n-1):
            b[i] = 2 * h * y[i] * (x[i]**3 - 1.5)

    # multiplying inverse A by b
    u = np.matmul(np.linalg.inv(A), b)
    
    # plotting
    plt.plot(x, u, 'b-', label = 'Numeric')
```

```python
def plotar():
    for c in range(1, 4):
        n = 10
        numeric_function(n**c)
        analytic_function()
        plt.xlabel('Axis X')
        plt.ylabel('Axis Y')
        plt.title(f'Exercise List 3: h = {n**c}')
        plt.legend(loc = 'upper left')
        plt.grid()
        plt.show()
plotar()
```
