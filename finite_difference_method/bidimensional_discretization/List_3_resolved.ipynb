```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def analytic_function(n):
    x = np.linspace(0, 2, n) # creating the x axis
    y = np.exp(x**4/4 - 1.5*x) # calculating y(x)
    plt.plot(x, y, 'r--', label = 'Analytic') # plotting
```

```python
def numeric_function(n): 
    # defining range
    start = 0
    final = 2
    # defining scope
    h = (final-start)/n
    # creating the x axis
    x = np.linspace(start , final, n)
    # creating matrix & vector
    A = np.zeros([n, n])
    b = np.zeros(n)
    # defining knowns points
    A[0,0] = 1
    b[-1] = 2.72
    b[0] = 1
    # setting the A matrix
    for i in range(1, n):
        A[i , i-1] = -1
        A[i, i] = 1
        b[i] = 2 * h * b[i-1] * (x[i]**3 - 1.5)
    # multiplying inverse A by b
    u = np.linalg.solve(A, b)
    # plotting
    plt.plot(x, u, 'b-', label = 'Numeric')
```

```python
def plotar():
    # lazy way to plot
    for c in range(1, 4):
        n = 10
        u = numeric_function(n**c)
        analytic_function(n**c)
        plt.xlabel('Axis X')
        plt.ylabel('Axis Y')
        plt.title(f'Exercise List 3: h = {n**c}')
        plt.legend(loc = 'upper left')
        plt.grid()
        plt.show()
```

```python
plotar()
```
