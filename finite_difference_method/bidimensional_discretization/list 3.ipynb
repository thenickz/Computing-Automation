```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def analytic_function():
    xx = np.linspace(0, 2)
    y = np.exp(np.power(xx, 4)/4 - 1.58*xx)
    plt.plot(xx, y, 'r--')
```

```python
def numeric_function2(n): 

    a = 0
    b = 2
    h = (b-a)/(n-1)
    x = np.linspace(a , 2, n)
    A = np.zeros([n, n])
    b = np.zeros(n)
    A[0,0] = 1
    b[0] = 2.72

    for i in range(1, n):
        A[i , i-1] = -1
        A[i, i] = -2
        b[i] = h * (np.exp(np.power(x[i], 4)/4 - 1.58*x[i]))*(x[i]**3 - 1.5)

    A[n-1,n-1] = 2
    b[-1] = 2.72
    u = A/b
    plt.plot(x, u,'b-')
```

```python

numeric_function2(10)
analytic_function()
plt.grid()
plt.show()

numeric_function2(100)
analytic_function()
plt.grid()
plt.show()

numeric_function2(1000)
analytic_function()
plt.grid()
plt.show()
```
