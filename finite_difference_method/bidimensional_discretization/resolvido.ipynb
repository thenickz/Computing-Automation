```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def analytic_function():
    x = np.linspace(0, 2, 1000)
    y = np.exp(x**4/4 - 1.5*x)
    plt.plot(x, y, 'r--.', label = 'Analytic', linewidth=1)
```

```python
def numeric_function(n):
    # defining range
    start = 0
    final = 2
    
    h = (final-start)/n
    
    x = np.linspace(start , final, n)

    A = np.zeros([n, n])
    b = np.zeros(n)
    y = np.zeros(n)
    b[-1] = 2.72
    b[0] = 1
    
    y[1:] = 1
    y[-1] = 2.72
    y[0] = 1

    for i in range(1, n-1):
        A[i, i+1] = 1
        A[i , i-1] = -1
    
    A[0,0] = 1
    A[n-1,n-1] = 1
    print(A)
    print(np.linalg.det(A))
    for j in range(1000):
        for i in range(1, n-1):
            b[i] = 2 * h * y[i] * (x[i]**3 - 1.5)
        y = np.linalg.solve(A, b)

    u = np.matmul(np.linalg.inv(A), b)

    plt.plot(x, u, 'b', label = 'Numeric', linewidth=1)

n = 10
for c in range(1, 4):
    numeric_function(n**c)
    analytic_function()
    plt.xlabel('Axis X')
    plt.ylabel('Axis Y')
    plt.title(f'Exercise List 3: h = {n**c}')
    plt.legend(loc = 'upper left')
    plt.grid()
    plt.show()
```
